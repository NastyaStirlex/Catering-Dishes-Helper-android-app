```python
def get_reviews(id):


    url = 'https://public-api.reviews.2gis.com/2.0/branches/'+f'{id}'+'/reviews?limit=12&is_advertiser=true&fields=meta.providers,meta.branch_rating,meta.branch_reviews_count,meta.total_count,reviews.hiding_reason,reviews.is_verified&without_my_first_review=false&rated=true&sort_by=date_edited&key=37c04fe6-a560-4549-b459-02309cf643ad&locale=ru_RU'
    print(url)
    response = (requests.get(url)).json()    
    items = []
    for element in response['reviews']:
        # print(element['result']['name'] + ' (' + str(element['resultScore']) + ')')
        items.append(element)
    return items
```
