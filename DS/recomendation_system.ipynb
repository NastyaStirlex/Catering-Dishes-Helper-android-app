```python

itemsfirst.duplicated().sum()
itemsfirst.drop_duplicates(inplace=True)

itemsfirst.isnull().sum()
itemsfirst.dropna(how='any',inplace=True)
```

```python
import nltk
from nltk.corpus import stopwords
import string
nltk.download('stopwords')
```

```python
itemsfirst["reviews_y"] = itemsfirst["reviews_y"].str.lower()

PUNCT_TO_REMOVE = string.punctuation
def remove_punctuation(text):
    """custom function to remove the punctuation"""
    return text.translate(str.maketrans('', '', PUNCT_TO_REMOVE))

itemsfirst["reviews_y"] = itemsfirst["reviews_y"].apply(lambda text: remove_punctuation(text))

STOPWORDS = set(stopwords.words('russian'))
def remove_stopwords(text):
    return " ".join([word for word in str(text).split() if word not in STOPWORDS])

itemsfirst["reviews_y"] = itemsfirst["reviews_y"].apply(lambda text: remove_stopwords(text))
itemsfirst[['reviews_y', 'cuisines']].sample(5)
```

```python
restaurant_names = list(itemsfirst['name'].unique())
def get_top_words(column, top_nu_of_words, nu_of_word):
    vec = CountVectorizer(ngram_range= nu_of_word, stop_words='russian')
    bag_of_words = vec.fit_transform(column)
    sum_words = bag_of_words.sum(axis=0)
    words_freq = [(word, sum_words[0, idx]) for word, idx in vec.vocabulary_.items()]
    words_freq =sorted(words_freq, key = lambda x: x[1], reverse=True)
    return words_freq[:top_nu_of_words]

df_percent = itemsfirst.sample(frac=0.5)
```

Выполняем предварительную обработку и создание матрицы TF-IDF и матрицы сходства косинусной, которую используем для поиска похожих отзывов, чтобы сделать рекомендации на основе текста отзывов.

```python
from nltk.corpus import stopwords
srword = nltk.corpus.stopwords.words('russian')
df_percent.set_index('name', inplace=True)
indices = pd.Series(df_percent.index)

tfidf = TfidfVectorizer(analyzer='word', ngram_range=(1, 2), min_df=0, stop_words=srword)
tfidf_matrix = tfidf.fit_transform(df_percent['reviews_y'])

cosine_similarities = linear_kernel(tfidf_matrix, tfidf_matrix)
```

```python
cosine_similarities.shape
```

```python
def recommend(name, cosine_similarities = cosine_similarities):
    recommend_restaurant = []
    idx = indices[indices == name].index[0]
    score_series = pd.Series(cosine_similarities[idx]).sort_values(ascending=False)
    top51_indexes = list(score_series.iloc[0:51].index)
    for each in top51_indexes:
        recommend_restaurant.append(list(df_percent.index)[each])
        df_new = pd.DataFrame(columns=['cuisines', 'general_rating', 'avg_price'])
    for each in recommend_restaurant:
        df_new = df_new.append(pd.DataFrame(df_percent[['cuisines', 'general_rating', 'avg_price']][df_percent.index == each].sample()))
    df_new = df_new.drop_duplicates(subset=['cuisines', 'general_rating', 'avg_price'], keep=False)
    df_new = df_new.head(15)
    print('Топ заведений с похожими отзывам: %d, %s' % (len(df_new), name))

    return df_new
recommend('Ватан, кафе')
```
