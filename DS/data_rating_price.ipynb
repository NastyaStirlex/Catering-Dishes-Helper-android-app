```python
import json
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import sys
import os
import urllib
import requests
```

```python
def get_items_page(page):

    key = 'ruzfsw2473'
    query = 'поесть'
    typev = 'branch'
    city_id = '422835235324469' 
    items = 'items.reviews,items.attribute_groups'
    service_url = 'https://catalog.api.2gis.com/3.0/items'
    params = {
        'q': query,
        'type': typev,
        'page': page,
        'city_id': city_id,
        'key': key,
        'fields' : items
        

    }
    url = service_url + '?' + urllib.parse.urlencode(params)
    print(url)
    response = (requests.get(url)).json()    
    items = []
    for element in response['result']['items']:
        # print(element['result']['name'] + ' (' + str(element['resultScore']) + ')')
        items.append(element)
    return items
def get_price_rating(items):
    ads = []
    for element in items['attribute_groups']:
      # print((element[0]['attributes'][0]['name'].replace('\xa0', '').split(' ')[-1]).replace('₽', ''))
        q = (element[0]['attributes'][0]['name'].replace('\xa0', '').split(' ')[-1]).replace('₽', '')
        # print(q)
        if q.isdigit():
            ads.append(q)
        else:
            ads.append(0)
    items['avg_price'] = ads
    df = []
    for element in items['reviews']:
        if 'general_rating' in element:
            df.append(element['general_rating'])
        else:
            df.append(0)
    items['general_rating'] = df
    return items
```

```python
dfs = {}
for i in range(1, 6):
    df_name = f"item{i}"
    dfs[df_name] = pd.DataFrame.from_dict(get_items_page(i))
pd = pd.concat(dfs, axis=0, ignore_index=True)
pd = get_price_rating(pd)
pd
```
